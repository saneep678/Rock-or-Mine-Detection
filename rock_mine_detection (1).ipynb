**Prepare Data for Model Training**

```python
# Separate features and target variable
X = df.drop(columns=60, axis=1)  # Features (all columns except last)
Y = df[60]  # Target variable (last column)
```

```python
print(X)
print(Y)
```

**Training and Testing Data**

```python
# Split dataset into training (90%) and testing (10%)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
# Check dataset shapes
print("Training Data Shape:", X_train.shape, Y_train.shape)
print("Testing Data Shape:", X_test.shape, Y_test.shape)
```

```python
print(X_train)
print(Y_train)
```

**Apply Feature Scaling**

```python
# Standardizing the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Check if scaling worked
print("X_train_scaled shape:", X_train_scaled.shape)
```

**Model Training Using Logistic Regression**

```python
# Train Logistic Regression
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train_scaled, Y_train)
```

```python
# Predictions
log_train_acc = accuracy_score(Y_train, log_model.predict(X_train_scaled))
log_test_acc = accuracy_score(Y_test, log_model.predict(X_test_scaled))
```

```python
# Print accuracy
print("Logistic Regression -> Train Accuracy:", log_train_acc, "| Test Accuracy:", log_test_acc)
```

**Model Training Using Random Forest**

```python
# Train Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=1)
rf_model.fit(X_train, Y_train)  # No scaling needed for RF
```

```python
# Predictions
rf_train_acc = accuracy_score(Y_train, rf_model.predict(X_train))
rf_test_acc = accuracy_score(Y_test, rf_model.predict(X_test))
```

```python
# Print accuracy
print("Random Forest -> Train Accuracy:", rf_train_acc, "| Test Accuracy:", rf_test_acc)
```

**Model Evaluation**

```python
# Confusion Matrix for Logistic Regression
log_cm = confusion_matrix(Y_test, log_model.predict(X_test_scaled))
sns.heatmap(log_cm, annot=True, fmt="d", cmap="Blues")
plt.title("Logistic Regression - Confusion Matrix")
plt.show()
```

```python
# Classification Report for Logistic Regression
print("Logistic Regression Report:\n", classification_report(Y_test, log_model.predict(X_test_scaled)))
```

```python
# Confusion Matrix for Random Forest
rf_cm = confusion_matrix(Y_test, rf_model.predict(X_test))
sns.heatmap(rf_cm, annot=True, fmt="d", cmap="Greens")
plt.title("Random Forest - Confusion Matrix")
plt.show()
```

```python
# Classification Report for Random Forest
print("Random Forest Report:\n", classification_report(Y_test, rf_model.predict(X_test)))
```
